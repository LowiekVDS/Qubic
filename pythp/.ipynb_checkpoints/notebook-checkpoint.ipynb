```python
import numpy as np
from board import *
from randomplayer import RandomPlayer
from minmaxplayer import MinMaxPlayer
from strategicplayer import StrategicPlayer
from collections import deque
from dqnplayer import DQNPlayer, agent

# Setup of board
b = Board()

p1 = DQNPlayer(b, CROSS)
p2 = RandomPlayer(b, NOUGHT)

nr_of_games = 1000

steps_to_update_target_model = 0

winners = deque([], maxlen=10000)

game_buffer = []

for i in range(nr_of_games):

    # Reset board        
    b.clear()
    nr_of_wildcards = np.random.randint(0,10)
    b.add_wildcards(nr_of_wildcards)
    p1.ok = False

    if i % 100 == 0:
        print(f"Iteration {i}/{nr_of_games}")
        for i in range(1, 4):
            print(f"Wins {i}: {winners.count(i) / max(len(winners),1)}")


    while b.state == ACTIVE:
        steps_to_update_target_model += 1

        # Make moves
        p1.move()
        p2.move()

    # Store game
    if p1.ok:
        game_buffer.append(p1.temp_game_buffer)
    p1.temp_game_buffer = []

    winners.append(b.state)

# Do some analysis
```

```python
dic = {}
print(len(game_buffer))
for i, game in enumerate(game_buffer):

    if len(game) == 0:
        continue
    
    action, state, trans = game[-1]
    state = np.array(state, dtype=int)
 
    key = hash(tuple(list(state[64:])))
    if key not in dic.keys():
        dic[key] = action
```

```python
import tensorflow as tf
from random import randint
import board
import numpy as np
import json
import tensorflow as tf
import keras
from keras.layers import Dense, Conv2D, Flatten, Conv3D
from collections import deque
import random
def agent():
    learning_rate = 0.001
    init = tf.keras.initializers.HeUniform()
    model = keras.Sequential()
    #model.add(Conv3D(2, kernel_size=2, activation='relu', input_shape=(4,4,4,2)))
    
   # model.add(Conv2D(1, kernel_size=4, activation='relu'))
    #model.add(Flatten())
    model.add(keras.layers.Dense(64, input_shape=(128,), activation='relu', kernel_initializer=init))
    model.add(keras.layers.Dense(64, activation='relu', kernel_initializer=init))
    model.add(keras.layers.Dense(64, activation='softmax', kernel_initializer=init))
    model.compile(loss=tf.keras.losses.Huber(), optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate), metrics=['accuracy'])

    return model
```

```python
X = []
Y = []

dic = {}

complete_state = np.zeros(64)

states = []

def to_board(state):
    board = np.zeros((4,4,4), dtype=int)
    
    X = state[:64].reshape((4,4,4)) - state[64:].reshape((4,4,4))
    X[X < 0] = 0
    
    Y = state[64:].reshape((4,4,4)) - state[:64].reshape((4,4,4))
    Y[Y < 0] = 0
    
    wild = state[64:].reshape((4,4,4)) + state[:64].reshape((4,4,4))
    wild[wild < 2] = 0
    
    board = X * CROSS + Y * NOUGHT + wild * WILDCARD
    
    return board
    
def make_board_str(board):
    res = ''
    res += "=========   =========   =========   =========\n"#.format(self.state, self.line)
    for i in range(4):
        for j in range(4):
            res += '|'
            for k in range(4):

                
                pos = board[i, j, k]

                if pos == CROSS:
                    res += 'X'
                elif pos == NOUGHT:
                    res += 'O'
                elif pos == WILDCARD:
                    res += '*'
                else:
                    res += ' '

                

                res += '|'
            res += '   '
        res += '\n'
    res += "=========   =========   =========   =========\n"
    return res


for i, game in enumerate(game_buffer):
    
    action, state, trans = game[-1]
    state = np.array(state, dtype=int)
    complete_state += state[:64]
    
    board = to_board(state)
    #print(state)
    #print(make_board_str(board))
     
    #print(state[:64].reshape((4,4,4)))
    #print(state[64:].reshape((4,4,4)))
    #print(action)
    #break
    
    #state = np.array(state, dtype=int)
    #print(state[:64].reshape((4,4,4)), state[64:].reshape((4,4,4)), action)
    #print((np.array(state[:64],dtype=int) - np.array(state[64:], dtype=int)).reshape((4,4,4)), action)
    #print(len(state[:64]))
        #amount += 1
        #key = hash(tuple(list(state[:64])))
        #if key in dic.keys():
        #    continue

        #dic[key] = action
    #continue
    for action, state, trans in game:
        
        key = hash(tuple(list(state[64:])))
        if key not in dic.keys():
            dic[key] = action
        
        if list(state) in states:
            continue
        
        states.append(list(state))
        
        X.append(state.reshape((128,1)))
        move = np.zeros(64, dtype=bool)
        move[action[0] * 16 + action[1] * 4 + action[2]] = True
        Y.append(move)

#print(complete_state)
Xtest = np.asarray(X)
Ytest = np.asarray(Y)

print(Xtest.shape,Ytest.shape)
#print(amount)
print(dic)
print(len(dic))
print(complete_state)
```

```python

model = agent()
model.summary()
```

```python
model.fit(Xtest, Ytest, epochs=200, use_multiprocessing=True, batch_size=32)
```
